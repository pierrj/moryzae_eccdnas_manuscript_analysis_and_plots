```python
import csv
import sys
```

```python
ltr_file = str(sys.argv[1])
eccs_file = str(sys.argv[2])
```

```python
ltr_file = 'tmp.LTR.PYRET.G3_1A.filtered.bed'
eccs_file = 'G3_1A.ecc_caller_out.genomebrowser.bed'
```

```python
with open(ltr_file, newline = '') as file:
    file_reader = csv.reader(file, delimiter = '\t')
    ltr_list = [[str(row[0]), int(row[1]), int(row[2]), str(row[3])] for row in file_reader]
    
with open(eccs_file, newline = '') as file:
    file_reader = csv.reader(file, delimiter = '\t')
    eccs_list = [[str(row[0]), int(row[1]), int(row[2])] for row in file_reader]
```

```python
print(ltr_list[0])
```

```python
putative_ltr_eccs = []
for ecc in eccs_list:
    ltr_int_list = []
    for ltr in ltr_list:
        if ltr[0] == ecc[0]:
            if (ecc[1] > ltr[1] and ecc[1] < ltr[2]) or (ecc[2] > ltr[1] and ecc[2] < ltr[2]):
                ltr_int_list.append('yo')
    if len(ltr_int_list) > 1:
        putative_ltr_eccs.append(ecc)
```

```python
for ecc in putative_ltr_eccs:
    print(ecc[0]+':'+str(ecc[1])+'-'+str(ecc[2]))
```
